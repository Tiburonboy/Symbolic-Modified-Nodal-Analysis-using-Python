## Numeric solution
Substitue the element values into the equations and solve for unknown node voltages and currents.

````python
# enter the element values
equ1a = equ.subs({
    R1:5.000000,
    R2:2.000000,
    R5:3.000000,
    R6:5.000000,
    R8:8.000000,
    R10:9.000000,
    R11:4.000000,
    R12:2.000000,
    R14:10.000000,
    R15:5.000000,
    V1:10.000000,
    V2:2.000000,
    V3:3.000000,
    V4:5.000000,
    I1:3.000000,
    I2:1.000000,
    I3:2.000000,
    I4:2.000000,
    R7:10.000000,
    R16:6.000000,
    R3:10.000000,
    R4:3.000000})

equ1a  # display the equations
````

````python
ans = solve(equ1a,X)
ans
````

````python
for name, value in ans.items():
    print('{:5s}: {:11.5f}'.format(str(name),float(value)))
    #print(name,value)
````
